```python
def read_xlsx_file(file_path, rows_to_skip=0):
    # Load the workbook
    workbook = openpyxl.load_workbook(file_path)

    # Assuming there is only one sheet in the workbook
    sheet = workbook.active

    lines = []

    # Iterate through rows
    for row in sheet.iter_rows(min_row=rows_to_skip + 1, values_only=True):
        
        # if row[0] == '#':
        #    continue

        lines.append(row)
        # print(row)

    # Close the workbook when done
    workbook.close()

    return lines

# Replace 'your_file.xlsx' with the path to your actual XLSX file
file_path = './data/commonsense_homework.xlsx'
lines = read_xlsx_file(file_path)
```

```python

def get_synset_name(word_name, word_class, word_lexname):

    synsets = wn.synsets(word_name, pos=wn.NOUN)

    for synset in synsets:
        if synset.lexname() == word_lexname:
            return synset.name()
    return None
```

```python
def write_preprocessed(lines, tsv_file_path, blank_set=None):

    with open(tsv_file_path, 'w', newline='') as tsv_file:

        # Create a CSV writer with tab as the delimiter
        tsv_writer = csv.writer(tsv_file, delimiter='\t')

        pbar = tqdm(lines)
        for line in pbar:

            # bar.set_description(line[0])
            pbar.set_postfix(word=line[0])

            data = [line[0], line[1], line[2], line[3]]

            if line[0] != '#':

                synset_name = get_synset_name(
                    line[0].replace(' ', '_').lower(), 
                    line[1], 
                    line[2]
                )

                if blank_set is not None and synset_name is None:
                    blank_set.add(line[0])

                data[3] = synset_name

            tsv_writer.writerow(data)

output_path = "./data/file.tsv"
blank_set = set()
write_preprocessed(lines, output_path, blank_set=blank_set)
```
